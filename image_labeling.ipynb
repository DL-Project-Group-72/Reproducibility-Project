```python
import cv2 as cv
import os
```

```python
path = "./10_archive/images/"
images = sorted(os.listdir(path))

all_imgs = {}

for num, image_name in enumerate(images):
    print(num, ' from ', len(images))
    tokens = image_name.split('_')
    
    is_gt = tokens[2] == 'gt'
    img_id = tokens[0]
    iterations = int(tokens[1])

    if is_gt:
        text = f"GT|{tokens[0]}|-|-|"
    else:
        text = f"Pred|{img_id}|0.1|{iterations}|"
    

    cur_img = cv.imread(path + image_name)
    
#     cv.namedWindow("output", cv.WINDOW_NORMAL)
#     cv.resizeWindow("output", 600, 400)# Create window with freedom of dimensions                   # Read image
#     imS = cv.resize(cur_img, (960, 540))                # Resize image
    #cv.imshow("output", imS)      
    #cv.waitKey(0)
    #cv.destroyAllWindows()
    
    cv.putText(img=cur_img, text=text, org=(150, 250), fontFace=cv.FONT_HERSHEY_TRIPLEX, fontScale=5, color=(0, 255, 0),thickness=3)
#     cv.namedWindow("output", cv.WINDOW_NORMAL)
#     cv.resizeWindow("output", 600, 400)# Create window with freedom of dimensions                   # Read image
#     imS = cv.resize(cur_img, (960, 540))   
    #cv.imshow("output", imS)      
    #cv.waitKey(0)
    #cv.destroyAllWindows()
    
    if is_gt:
        saved_name = f"{img_id}_0"
    else:
        saved_name = f"{img_id}_{iterations}"
    
    cv.imwrite(f"./10_archive/labeled_images/{saved_name}.jpg", cur_img)
    
    
```
